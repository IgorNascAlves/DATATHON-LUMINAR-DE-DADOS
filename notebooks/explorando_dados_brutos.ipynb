Descobri que tem colunas por anos, vou separ por arquivo

```python
#regex para pegar nome de colunas iguais, removendo os 4 ultimos caracteres que é o ano
lista_colunas = df_pede_passos.columns.str.extract(r'(\w+)(?=_\d{4})')

print(f"Colunas totais {len(df_pede_passos.columns)}")
print(f"Colunas unicas {len(lista_colunas[0].unique())}")
```

Nem todas as colunas aparecem em todos os anos

```python
#check equals lines on lista_colunas
lista_colunas_com_duas = lista_colunas[0].value_counts()[lista_colunas[0].value_counts() == 2]
display(lista_colunas_com_duas)
lista_colunas_com_tres = lista_colunas[0].value_counts()[lista_colunas[0].value_counts() == 3]
display(lista_colunas_com_tres)
```

#### Separando dados por ano

```python
#create a new dataframe with columns names that ends with 2021
df_pede_passos_2021 = df_pede_passos.filter(regex='_2021$', axis=1)

df_pede_passos_2021.info()
```

```python
#create a new dataframe with columns names that ends with 2022
df_pede_passos_2022 = df_pede_passos.filter(regex='_2022$', axis=1)

df_pede_passos_2022.info()
```

```python
#create a new dataframe with columns names that ends with 2020
df_pede_passos_2020 = df_pede_passos.filter(regex='_2020$', axis=1)

df_pede_passos_2020.info()
```

```python
#create a new dataframe with columns names thats not ends with 2022 or 2021 or 2020
df_pede_passos_outros = df_pede_passos.filter(regex='^(?!.*_(2020|2021|2022)$).*$', axis=1)

df_pede_passos_outros.info()
```

```python
# save dataframes to csv
df_pede_passos_2020.to_csv('..\dados\dados_processados\pede_passos_2020.csv', sep=';', index=False)
df_pede_passos_2021.to_csv('..\dados\dados_processados\pede_passos_2021.csv', sep=';', index=False)
df_pede_passos_2022.to_csv('..\dados\dados_processados\pede_passos_2022.csv', sep=';', index=False)
```
